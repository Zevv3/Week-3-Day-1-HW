## Lambda Functions <br>
<p>Lambda functions... or "Anonymous Functions" are referring to inline functions with no name. The keyword lambda denotes the no name function, and executes within a single line. Without saving it to a variable; however, it is not able to be used, unless passed in either as a paramater or within list comprehension.<br>Written as "(keyword lambda) (one or more inputs) (colon) (function to be executed)"</p>

#### Syntax

```python
#regular function

def addTwo(num):
    return num +2

print(addTwo(4))

#Lambda syntax for defining a function without saving it to a variable (addTwo would be the variable in the previous case)
#Kind of like list comprehsnsion but for a function

print((lambda num: num + 2)(4))
```

#### Saving to a Variable

```python
#We can save lambdas as variables

f_test = lambda num: num + 2

print(f_test(4))
```

#### Multiple Inputs

```python
#Lambda function with multiple inputs with no variable

print((lambda x,y,z: x * y * z)(3,5,8))

#long way of doing this

def multiply_nums(x,y,z):
    return x * y * z

print(multiply_nums(3,5,8))

#Of course, you can also save these to variables

num_test = lambda x,y,z: x * y * z
print(num_test(3,5,8))
```

#### Passing a Lambda into a Function

```python
def multiple(func, num):
    """
    func expects a lambda function
    and num expects an integer
    """
    return func(num)

print(multiple(lambda x: x * x, 4))
```

#### Returning a Lambda from a Function

```python
#Regular way of defining the fucntion

def multiply_test(num):
    return num * 4

print(multiply_test(2))

#Syntax for writing functions inside of functions

def return_func():
    test = 4
    def multiply(num):
        return num * 4
    return multiply

f_return = return_func()
print(f_return(2))

#this looks awful, so instead we use lambda functions inside functions

def return_lamb(a, b):
    return lambda c, d: c + d + a + b

r_return = return_lamb(3, 4)
print(r_return)
print(r_return(5,6))
```

```python
#another example using lambda function inside a function

def my_func(num1, num2):
    print(num1, num2)
    return lambda num3, num4: (num3 * num4) + (num1 * num2)

alex_var = my_func(1,2)
print(alex_var)
print(alex_var(3,4))

#you actually don't have to save to a variable:
print(my_func(1,2)(3,4))
```

#### If Statements within Lambdas

```python
# syntax for if stateents in lambdas 
# lambda x: True if (condition) else False

f_conditional = lambda num: num * 2 if num > 10 else num + 2

print(f_conditional(8))
print(f_conditional(10))
print(f_conditional(12))
```

```python
# example of if-elif-else using lambdas
#the backslash allows you to enter down a line and python reads it as a single line
alex_examp = lambda num: f"{num} is less than 5" if num < 5 else\
(f"{num} is greater than 5 and less than 10" if num < 10 else f"{num} is greater than 10")

print(alex_examp(3))
print(alex_examp(8))
print(alex_examp(12))
```

```python
#multiple elifs:
alex_examp = lambda num: f"{num} is less than 5" if num < 5 else\
(f"{num} is greater than 5 and less than 10" if num < 10 else\
 (f"{num} is greater than 10 and less than 20" if num < 20 else f"{num} is greater than 20"))

print(alex_examp(12))
print(alex_examp(22))
```

## Map <br>
<p>The map function allows you to iterate over an entire list while running a function on each item of the list. This is why the map function works well with lambda's, because it simplifies things and you write less lines of code.<br>The syntax for a map function is "map(function to be used, list to be used)"<br>However, you must be careful, as the map function returns a map object, not a list. To turn it into a list we use the list() type conversion.</p>

#### Syntax

```python
# syntax for utilizing map function
# map(functions, iterable(lists,dicts,tuples,sets,etc.))
# Normally the usage of map happens with a pre-defined function -- but we can use lambdas

#utilizing map function with a regular pre-defined user function

def squared(num1, num2):
    if num1 < 10 and num2 < 10:
        return num1 ** 2, num2 ** 2
    else:
        return num1, num2
    
numbers = [4, 11, 20, 3, 15, 20]
more_nums = [4, 10, 3, 2, 6]

squared_nums = list(map(squared, numbers, more_nums))
print(squared_nums)

# note that above, 20 in the list numbers doesn't run because the other list had already ended. It's like a 
# for loop that runs while there are items in the lists

```

#### Using Lambda's with Map

```python
# syntax for map & lambdas
# list(map(lambda x: (transformation on x), list (or other iterable)))
# usually happens in one line so it's cleaner code

squared_nums_again = list(map(lambda num1, num2: (num1 ** 2, num2 ** 2) if num1 < 10 and num2 < 10\
                             else (num1, num2), numbers, more_nums))
print(squared_nums_again)
```

## Filter() <br>
<p>Filter's are similar to the map function, where you're able to pass a function argument and a list argument and filter out something from the list based on the conditions passed. Similar to the map function, it returns a filter object, so you need to type convert it to a list()</p>

#### Syntax

```python
#syntax for filter:
#list(filter(function, iterable))

names = ['Chris', 'Steve', 'Perry', 'Kelsey']

def y_names(name):
    if name[-1].lower() == 'y':
        return True
    else:
        return False
    
new_names_list = list(filter(y_names, names))
print(new_names_list)
```

#### Using Lambda's with Filter()

```python
new_names_lamb = list(filter(lambda name: True if name[-1].lower() == 'y' else False, names))
print(new_names_lamb)
```

## Reduce() <br>
<p>Be very careful when using this function, as of Python 3 it's been moved to the 'functools' library and no longer is a built-in function.<br>The creator of Python himself, says to just use a for loop instead.</p>

#### Syntax

```python
from functools import reduce

#syntax for reduce:
#reduce(functions, iterable)

list_1 = [2, 4, 6, 8, 10]

#user-function
def add_nums(num1,num2):
    return num1 + num2

result_add = reduce(add_nums, list_1)
print(result_add)

#use python tutor to see what the heck this does
#doing this with a for loop instead:
result = 0
for num in list_1:
    result += num
    
print(result)
```

#### Using Lambda's with Reduce()

```python
reduce_lamb = reduce(lambda num1, num2: num1 + num2, list_1)
print(reduce_lamb)
```

## Recursion <br>
<p>Recursion means that a function is calling itself, so it contanstly executes until a base case is reached. It will then push the returning values back up the chain until the function is complete. A prime example of recursion is computing factorials... such that 5! (factorial) is 5*4*3*2*1 which equals 120.</p>

#### Implementing a Base Case

```python
def add_nums(num):
    #first we want to set our base case
    if num <= 1:
        print("add_nums(1) = 1")
        return num
    else:
        print(f"add_nums({num} = {num} + add_nums{num - 1})")
        return num + add_nums(num - 1)
    
add_nums(5)
```

#### Writing a Factorial Function

```python
def factorial(num):
    #set our base case
    if num <= 1:
        return 1
    else:
        return num * factorial(num-1)

print(factorial(5))
```

#### In-Class Exercise #5 <br>
<p>Write a recursive function that subtracts all numbers to the argument given.</p>

```python
def sub_nums(num):
    if num <= 1:
        print("sub_nums(1) = 1")
        return num
    else:
        print(f"sub_nums({num}) = {num} - sub_nums({num - 1})")
        return num - sub_nums(num-1)
print(sub_nums(5))
```

```python
#recursion is like a tree. You keep going down the tree until you reach the base case (roots) which once
# you've reached the roots you can start growing the tree

#5 - 4 (3 - (2 -1)) = 3
#4 - (3 - (2 - 1)) = 2
#3 - (2 - 1)) = 2
#2 - (1-1) = 2
#1 is our base case (we can plant our roots)
```

## Generators <br>
<p>Generators are a type of iterable, like lists or tuples. They do not allow indexing, but they can still be iterated through with for loops. They are created using functions and the yield statement.</p>

#### Yield Keyword <br>
<p>The yield keyword denotes a generator, it doesn't return so it won't leave the function and reset all variables in the function scope, instead it yields the number back to the caller.</p>

```python
def my_range(stop, start, step = 2):
    while start < stop:
        yield start #yield keyword is what makes it a generator
        start += step
        
for i in my_range(20, 2):
    my_generator_value = i
    print(my_generator_value)
```

#### Infinite Generator

```python
# bad, never create infinite loops

def bad_range(stop, start, step=1):
    while start < stop:
        yield start
        #make sure you always increment your generators!
        
for i in bad_range(10,1):
    print(i)
```

#### In-Class Exercise #6 <br>
<p>Create a generator that takes a number argument and yields that number squared, then prints each number squared until zero is reached.</p>

```python
def my_range(stop, start, step = 1):
    while start > stop:
        result = start ** 2
        yield result
        print(result)
        start -= 1

for i in my_range(0, 5):
    print(i)
```

### Exercise #4 <br>
<p>Write a recursion function to perform the fibonacci sequence up to the number passed in.</p>

`Output for fib(5) => 
Iteration 0: 1
Iteration 1: 1
Iteration 2: 2
Iteration 3: 3
Iteration 4: 5
Iteration 5: 8`

```python
def fib(num):
    if num <= 1:
        return num
    else:
        return(fib(num-1) + fib(num-2))

num = int(input("How many numbers of the fibonacci sequence would you like to see? "))
for i in range(num):
    print(fib(i))
```
